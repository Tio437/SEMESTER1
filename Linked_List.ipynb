Singly Linked List

```python
#Node dalam linked list
class TaskNode:
  def __init__(self, task_name):
    self.task_name = task_name #Nama tugas
    self.next = None #Pointer

#Linked list untuk daftar tugas (task list)
class TaskList:
  def __init__(self):
    self.head = None #Tidak ada tugas pada awalnya (daftar kosong)

  def add_task(self, task_name): #Menambah tugas
    new_task = TaskNode(task_name)
    if not self.head:
      self.head = new_task #Jika list kosong, tugas pertama menjadi head
    else:
      current = self.head
      while current.next: #Mencari node terakhir dalam list
        current = current.next
      current.next = new_task #Menambah tugas di akhir list

  def show_task(self): #Menampilkan daftar tugas
    if not self.head:
      print("\nTidak ada tugas hari ini")
      return
    print("\nDaftar tugas hari ini:")
    current = self.head
    count = 1
    while current: #Perulangan tak hingga sampai list tugas ditampilkan semua
      print(f"{count}. {current.task_name}")
      current = current.next
      count += 1

#Program utama (main)
task_list = TaskList() #Membuat objek daftar tugas
num_task = int(input("Masukkan jumlah tugas hari ini: ")) #User memasukkan jumlah tugas

#Loop untuk memasukkan tugas ke linked list
for i in range(1, num_task + 1):
  task_name = input(f"Masukkan tugas ke-{i}: ")
  task_list.add_task(task_name)

#Menampilkan hasil (output)
task_list.show_task()
```

Doubly Linked List

```python
# Definisi Node dalam Doubly Linked List
class PageNode:
    def __init__(self, url):
        self.url = url  # URL halaman web
        self.next = None  # Pointer ke halaman berikutnya
        self.prev = None  # Pointer ke halaman sebelumnya

# Definisi Doubly Linked List untuk Riwayat Peramban
class BrowserHistory:
    def __init__(self):
        self.current = None  # Pointer ke halaman yang sedang dibuka

    def visit_page(self, url):  # Menambahkan halaman baru ke dalam riwayat
        new_page = PageNode(url)
        if not self.current:  # Jika riwayat kosong, halaman ini menjadi halaman pertama
            self.current = new_page
        else:
            new_page.prev = self.current
            self.current.next = new_page
            self.current = new_page  # Pindahkan posisi ke halaman baru

    def show_history(self): # Menampilkan riwayat kunjungan
        if not self.current:
            print("\nRiwayat kosong.")
            return
        print("\nRiwayat Peramban Web:")
        temp = self.current
        while temp.prev: # Kembali ke halaman pertama
            temp = temp.prev
        count = 1
        while temp:
            print(f"{count}. {temp.url}")
            temp = temp.next
            count += 1

    def go_back(self): # Navigasi mundur ke halaman sebelumnya
        if self.current and self.current.prev:
            self.current = self.current.prev
            print(f"\nAnda kembali ke: {self.current.url}")
        else:
            print("\nTidak bisa kembali, ini adalah halaman pertama!")

    def go_forward(self): # Navigasi maju ke halaman berikutnya
        if self.current and self.current.next:
            self.current = self.current.next
            print(f"\nAnda maju ke: {self.current.url}")
        else:
            print("\nTidak bisa maju, ini adalah halaman terakhir!")

# Program Utama (Main)

history = BrowserHistory() # Membuat objek riwayat peramban
num_pages = int(input("\nMasukkan jumlah halaman yang dikunjungi: ")) # Memasukkan jumlah halaman yang dikunjungi

# Loop untuk memasukkan URL halaman ke dalam Doubly Linked List
for i in range(1, num_pages + 1):
    url = input(f"Masukkan URL halaman ke-{i}: ")
    history.visit_page(url)

history.show_history() # Menampilkan riwayat yang telah dimasukkan

# Menu navigasi pengguna
while True:
    command = input("\nTekan 'b' untuk kembali, Tekan 'f' untuk maju, Tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()
    if command == 'b':
        history.go_back()
    elif command == 'f':
        history.go_forward()
    elif command == 'x':
        print("\nKeluar dari riwayat peramban.")
        break
    else:
        print("\nPerintah tidak dikenal.")
```

Circular Linked List

```python
# Definisi Node dalam Circular Linked List
class SongNode:
    def __init__(self, title):
        self.title = title  # Menyimpan judul lagu
        self.next = None  # Pointer ke lagu berikutnya

# Definisi Circular Linked List untuk Playlist Lagu
class Playlist:
    def __init__(self):
        self.current = None  # Lagu yang sedang diputar

    def add_song(self, title): # Menambahkan lagu ke dalam playlist
        new_song = SongNode(title)
        if not self.current:
            self.current = new_song
            self.current.next = self.current  # Menghubungkan ke dirinya sendiri (Circular)
        else:
            temp = self.current
            while temp.next != self.current:  # Mencari lagu terakhir
                temp = temp.next
            temp.next = new_song  # Menambahkan lagu di akhir
            new_song.next = self.current  # Menghubungkan kembali ke awal

    def show_playlist(self): # Menampilkan daftar lagu dalam playlist
        if not self.current:
            print("\nPlaylist kosong!")
            return
        print("\nPlaylist Lagu Anda:")
        temp = self.current
        count = 1
        while True:
            print(f"{count}. {temp.title}")
            temp = temp.next
            count += 1
            if temp == self.current:  # Berhenti jika sudah kembali ke awal
                break

    def next_song(self): # Memutar lagu berikutnya dalam playlist
        if self.current:
            self.current = self.current.next
            print(f"\nMemutar: {self.current.title}")

# Program Utama (Main)

playlist = Playlist() # Membuat objek playlist
num_songs = int(input("\nMasukkan jumlah lagu dalam playlist: ")) # Memasukkan jumlah lagu

# Loop untuk memasukkan lagu ke dalam Circular Linked List
for i in range(1, num_songs + 1):
    title = input(f"Masukkan judul lagu ke-{i}: ")
    playlist.add_song(title)

playlist.show_playlist() # Menampilkan daftar lagu

# Menu navigasi pengguna
while True:
    command = input("\nTekan 'n' untuk lagu berikutnya, Tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()
    if command == 'n':
        playlist.next_song()
    elif command == 'x':
        print("\nKeluar dari pemutar musik.")
        break
    else:
        print("\nPerintah tidak dikenal.")
```

Operasi Dasar Linked List

```python
# Definisi Node dalam Singly Linked List
class Node:
    def __init__(self, data):
        self.data = data  # Menyimpan nilai node
        self.next = None  # Pointer ke node berikutnya

# Definisi Singly Linked List dengan Berbagai Operasi
class LinkedList:
    def __init__(self):
        self.head = None  # Awalnya, daftar kosong

    # Menambahkan elemen di awal
    def add_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Menambahkan elemen di akhir
    def add_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Menambahkan elemen di posisi tertentu
    def add_at_position(self, position, data):
        if position < 0:
            print("\nPosisi tidak valid!")
            return
        new_node = Node(data)
        if position == 0:
            self.add_at_beginning(data)
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp:
                print("\nPosisi melebihi panjang daftar!")
                return
            temp = temp.next
        new_node.next = temp.next
        temp.next = new_node

    # Menghapus elemen di awal
    def delete_at_beginning(self):
        if not self.head:
            print("\nDaftar kosong, tidak ada yang bisa dihapus!")
            return
        self.head = self.head.next

    # Menghapus elemen di akhir
    def delete_at_end(self):
        if not self.head:
            print("\nDaftar kosong, tidak ada yang bisa dihapus!")
            return
        if not self.head.next:
            self.head = None
            return
        temp = self.head
        while temp.next.next:
            temp = temp.next
        temp.next = None

    # Menghapus elemen di posisi tertentu
    def delete_at_position(self, position):
        if not self.head:
            print("\nDaftar kosong, tidak ada yang bisa dihapus!")
            return
        if position == 0:
            self.delete_at_beginning()
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp.next:
                print("\nPosisi tidak ditemukan!")
                return
            temp = temp.next
        temp.next = temp.next.next

    # Mencari elemen dalam Linked List
    def search(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
                print(f"\nElemen {key} ditemukan dalam daftar!")
                return
            temp = temp.next
        print(f"\nElemen {key} tidak ditemukan dalam daftar.")

    # Membalik (reverse) Linked List
    def reverse(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev

    # Menampilkan Linked List
    def display(self):
        if not self.head:
            print("\nDaftar kosong!")
            return
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

# Program Utama (Main)

# Membuat objek Linked List
ll = LinkedList()

# Meminta pengguna memasukkan elemen awal
num_elements = int(input("Masukkan jumlah awal elemen dalam daftar: "))
for i in range(num_elements):
    data = int(input(f"Masukkan data ke-{i+1}: "))
    ll.add_at_end(data)

# Menu Operasi Linked List
while True:
    print("\n=== Pilih Operasi ===")
    print("1. Tambah elemen di awal")
    print("2. Tambah elemen di akhir")
    print("3. Tambah elemen di posisi tertentu")
    print("4. Hapus elemen di awal")
    print("5. Hapus elemen di akhir")
    print("6. Hapus elemen di posisi tertentu")
    print("7. Cari elemen dalam daftar")
    print("8. Balik elemen dalam daftar")
    print("9. Tampilkan daftar")
    print("0. Keluar")

    choice = int(input("\nMasukkan pilihan: "))

    if choice == 1:
        data = int(input("Masukkan data yang ingin ditambahkan di awal: "))
        ll.add_at_beginning(data)
    elif choice == 2:
        data = int(input("Masukkan data yang ingin ditambahkan di akhir: "))
        ll.add_at_end(data)
    elif choice == 3:
        position = int(input("Masukkan posisi untuk menambahkan data: "))
        data = int(input("Masukkan data: "))
        ll.add_at_position(position, data)
    elif choice == 4:
        ll.delete_at_beginning()
    elif choice == 5:
        ll.delete_at_end()
    elif choice == 6:
        position = int(input("Masukkan posisi elemen yang ingin dihapus: "))
        ll.delete_at_position(position)
    elif choice == 7:
        key = int(input("Masukkan elemen yang ingin dicari: "))
        ll.search(key)
    elif choice == 8:
        ll.reverse()
        print("\nLinked List berhasil dibalik!")
    elif choice == 9:
        ll.display()
    elif choice == 0:
        print("\nKeluar dari program.")
        break
    else:
        print("\nPilihan tidak valid, silakan coba lagi!")
```

```python
class NodeJadwal:
  def __init__(self, nama_tugas):
    self.nama_tugas = nama_tugas
    self.next = None

class JadwalHariJumat:
  def __init__(self):
    self.head = None

  def tambah_tugas(self, nama_tugas):
    tugas_baru = NodeJadwal(nama_tugas)
    if not self.head:
      self.head = tugas_baru
    else:
      current = self.head
      while current.next:
        current = current.next
      current.next = tugas_baru

  def show_task(self):
    if not self.head:
      print("\nTidak ada tugas di hari Jum'at")
      return
    print("\nDaftar tugas hari Jum'at:")
    current = self.head
    count = 1
    while current:
      print(f"{count}. {current.nama_tugas}")
      current = current.next
      count += 1

jadwal_hari_jumat = JadwalHariJumat()
jumlah_tugas = int(input("Jumlah tugas hari Jum'at (digit terakhir NPM): "))

for i in range(1, (jumlah_tugas %10) + 1):
  nama_tugas = input(f"Masukkan tugas: ")
  jadwal_hari_jumat.tambah_tugas(nama_tugas)

jadwal_hari_jumat.show_task()
```

```python
class NodeHistory:
    def __init__(self, folder):
        self.folder = folder
        self.next = None
        self.prev = None

class FolderHistory:
    def __init__(self):
        self.current = None

    def visit_folder(self, folder):
        folder_baru = NodeHistory(folder)
        if not self.current:
            self.current = folder_baru
        else:
            folder_baru.prev = self.current
            self.current.next = folder_baru
            self.current = folder_baru

    def show_history(self):
        if not self.current:
            print("\nRiwayat kosong.")
            return
        print("\nRiwayat Folder:")
        temp = self.current
        while temp.prev:
            temp = temp.prev
        count = 1
        while temp:
            print(f"{count}. {temp.folder}")
            temp = temp.next
            count += 1

    def mundur(self):
        if self.current and self.current.prev:
            self.current = self.current.prev
            print(f"\nAnda kembali ke: {self.current.folder}")
        else:
            print("\nTidak bisa kembali, ini adalah folder pertama!")

    def maju(self):
        if self.current and self.current.next:
            self.current = self.current.next
            print(f"\nAnda maju ke: {self.current.folder}")
        else:
            print("\nTidak bisa maju, ini adalah folder terakhir!")

history = FolderHistory()
jumlah_folder = int(input("\nJumlah folder yang dikunjungi (digit terakhir NPM): "))

for i in range(1, (jumlah_folder % 10) + 1):
    folder = input(f"Masukkan folder ke-{i}: ")
    history.visit_folder(folder)

history.show_history()

while True:
    command = input("\nTekan 'b' untuk kembali, Tekan 'f' untuk maju, Tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()
    if command == 'b':
        history.mundur()
    elif command == 'f':
        history.maju()
    elif command == 'x':
        print("\nKeluar dari riwayat file explorer.")
        break
    else:
        print("\nPerintah tidak dikenal.")
```
