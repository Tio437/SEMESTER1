```python
def luas_persegi(sisi):
    return sisi * sisi

def luas_segitiga(alas, tinggi):
    return 0.5 * alas * tinggi

def luas_lingkaran(jari_jari):
    return 3.14 * jari_jari * jari_jari

def pilih ():
  while True:
    print("Pilih bentuk yang ingin dihitung luasnya:")
    print("1. Persegi")
    print("2. Segitiga")
    print("3. Lingkaran")
    print("4. selesai")
    try:
      pilihan = int(input("Masukkan nomor pilihan Anda: "))
    except ValueError:
      print("erorrrrrrrrrrrrrrr")
      continue
    if pilihan == 1:
      try:
        sisi = float(input("Masukkan panjang sisi persegi: "))
      except ValueError:
        print("hanya boleh memasukkan angka")
        continue
      luas = luas_persegi(sisi)
      print(f"Luas persegi dengan sisi {sisi} adalah {luas}")
    elif pilihan == 2:
      try :
        alas = float(input("Masukkan panjang alas segitiga: "))
        tinggi = float(input("Masukkan tinggi segitiga: "))
      except valueeror :
        print("hanya boleh memasukkan angka")
      luas = luas_segitiga(alas, tinggi)
      print(f"Luas segitiga dengan alas {alas} dan tinggi {tinggi} adalah {luas}")
    elif pilihan == 3:
      try :
        jari_jari = float(input("Masukkan jari-jari lingkaran: "))
        luas = luas_lingkaran(jari_jari)
      except valueeror :
         print("hanya boleh memasukan angka")
      print(f"Luas lingkaran dengan jari-jari {jari_jari} adalah {luas}")
    elif pilihan == 4:
      print("Terima kasih! Program selesai.")
      break
    else:
      print("Pilihan tidak ada. Silakan coba lagi.")
pilih()
```

```python
try:
  npm = int(input("Masukkan NPM Anda: "))
  genap = 0
  ganjil = 0
  for nomor in npm :
      nomor=int(nomor)
      if nomor % 2 == 0 :
          genap += 1
      else :
          ganjil += 1
  print("Jumplah bilangan Ganjil =",ganjil)
  print("Jumplah bilangan Genap =",genap)
except ValueError:
  print("NPM harus berupa angka.")
```
