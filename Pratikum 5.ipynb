```python
num1 = int(input("Angka Pertama : "))
num2 = int(input("Angka Kedua : "))
num3 = int(input("Angka Ketiga : "))
if num1 > num2 and num1 > num3:
    print(num1,"Nilai paling Terbesar")
elif num2 > num1 and num2 > num3:
    print(num2,"Nilai paling Terbesar")
else:
    print(num3,"Nilai paling Terbesar")
```

```python
jenis_member = input("Member atau non member : ")
if jenis_member =='member':
   member = int(input("Masukan total belanja : "))
   if member >= 200000:
    member = member - (member  * (4/100+2/100))
    print("mendapatkan Diskon Belanja lebih dari 200000 sebesar 4% dan tambahan diskon member 2% =Rp",member)
   elif member >= 100000:
    member = member - (member * (3/100+2/100))
    print("mendapatkan Diskon Belanja lebih dari 100000 sebesar 3% dan tambahan diskon member 2% =Rp",member)
   else:
    print("mendapatkan diskon member 2% =Rp",member-(member*2/100))
elif jenis_member == 'non member':
    non_member = int(input("Masukan total belanja : "))
    if non_member >= 200000:
        non_member = non_member - (non_member * 4/100)
        print("mendapatkan Diskon Belanja lebih dari 200000 sebesar 4% =Rp",non_member)
    elif non_member >= 100000:
        non_member = non_member - (non_member * 3/100)
        print("mendapatkan Diskon Belanja lebih dari 100000 sebesar 3% =Rp",non_member)
    else:
        print("total",non_member)
else:
    print("masukan member atau non member")
```
