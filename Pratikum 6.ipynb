```python
n = int(input("Masukkan nomor baris segitiga Pascal: "))

baris = [1]

for _ in range(n - 1):
    baris_baru = [1]
    for i in range(len(baris) - 1):
        baris_baru.append(baris[i] + baris[i + 1])
    baris_baru.append(1)
    baris = baris_baru

hasil = sum(baris)
print(f"Jumlah angka pada baris ke-{n} segitiga Pascal adalah: {hasil}")
print("".join (map(str,baris)))
```
