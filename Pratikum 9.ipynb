```python
data_diri = {
        "Nama         "  : "Tio Restu Pambudi",
        "Romble       "  : "02",
        "NPM          "  : "2440506076",
        "Tanggal Lahir"  : "18 Juni 2004",
        "Alamat       "  : "Magelang",
        "Hobi"           : {
                           "1":"Investasi",
                           "2":"Mencari uang",
                           "3":"Mengaji"
        }
}

for key, value in data_diri.items():
    if key == "Hobi":
        print(f"{key}          :")
        for hobi in value.values():
            print(f"               - {hobi}")
    else:
        print(f"{key} : {value}")
```
