```python
npm = input("masukan NPM Anda :  ")
genap = 0
ganjil = 0
for nomor in npm :
    nomor=int(nomor)
    if nomor % 2 == 0 :
        genap += 1
    else :
        ganjil += 1
print("Jumplah bilangan Ganjil =",ganjil)
print("Jumplah bilangan Genap =",genap)
```

```python
b = int(input("Masukkan jumlah baris: "))
for i in range(1, b + 1):
    for j in range(1, i + 1):
        print(i*j, end=" ")
    print()
```
