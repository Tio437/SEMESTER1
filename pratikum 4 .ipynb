WELKYY EXERCISE 3

```python
nilai = int(input("Masukkan Nilai "))

if 80 <= nilai <= 100:
  print("Grade A")
elif 77.50 <= nilai <= 79.99:
  print("Grade A-")
elif 75.50 <= nilai <= 77.49:
  print("Grade AB")
elif 72.50 <= nilai <= 74.49:
  print("Grade B+")
elif 70 <= nilai <= 72.49:
  print("Grade B")
elif 67.50 <= nilai <= 69.99:
  print("Grade B-")
elif 65 <= nilai <= 67.49:
  print("Grade BC")
elif 62.50 <= nilai <= 64.99:
  print("Grade C+")
elif 60 <= nilai <= 62.49:
  print("Grade C")
elif 55 <= nilai <= 59.99:
  print("Grade C-")
elif 50 <= nilai <= 54.99:
  print("Grade CD")
elif 45 <= nilai <= 49.99:
  print("Grade D+")
elif 40 <= nilai <= 44.99:
  print("Grade D")
else:
  print("Grade E")
#
```

```python
nilai = float(input("Masukkan Nilai "))

if nilai <= 100 and nilai >= 80:
  print("Grade A")
elif nilai <= 79.99 and nilai >= 77.50:
  print("Grade A-")
elif nilai <= 77.49 and nilai >= 75:
  print("Grade AB")
elif nilai <= 74.99 and nilai >= 72.50:
  print("Grade B+")
elif nilai <= 72.49 and nilai >= 70:
  print("Grade B")
elif nilai <= 69.99 and nilai >= 67.50:
  print("Grade B-")
elif nilai <= 67.49 and nilai >= 65:
  print("Grade BC")
elif nilai <= 64.99 and nilai >= 62.50:
  print("Grade C+")
elif nilai <= 62.49 and nilai >= 60:
  print("Grade C")
elif nilai <= 59.99 and nilai >= 55:
  print("Grade C-")
elif nilai <= 54.99 and nilai >= 50:
  print("Grade CD")
elif nilai <= 49.99 and nilai >= 45:
  print("Grade D+")
elif nilai <= 44.99 and nilai >= 40:
  print("Grade D")
else:
  print("Grade E")
```
